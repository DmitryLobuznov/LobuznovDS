```python
def create_dict_to_fillna(column):
    # –°–æ–∑–¥–∏–º —Å–ª–æ–≤–∞—Ä—å –ø–æ employment_type –¥–ª—è –∑–∞–ø–æ–ª–Ω–µ–Ω–∏—è —Å—Ç–æ–ª–±—Ü–∞ `column`.
    dict_fillna = {
        '—Å–æ—Ç—Ä—É–¥–Ω–∏–∫' : df[df['employment_type']=='—Å–æ—Ç—Ä—É–¥–Ω–∏–∫'][column].median(),
        '–ø–µ–Ω—Å–∏–æ–Ω–µ—Ä' : df[df['employment_type']=='–ø–µ–Ω—Å–∏–æ–Ω–µ—Ä'][column].median(),
        '–∫–æ–º–ø–∞–Ω—å–æ–Ω' : df[df['employment_type']=='–∫–æ–º–ø–∞–Ω—å–æ–Ω'][column].median(),
        '–≥–æ—Å—Å–ª—É–∂–∞—â–∏–π' : df[df['employment_type']=='–≥–æ—Å—Å–ª—É–∂–∞—â–∏–π'][column].median(),
        '–±–µ–∑—Ä–∞–±–æ—Ç–Ω—ã–π' : df[df['employment_type']=='–±–µ–∑—Ä–∞–±–æ—Ç–Ω—ã–π'][column].median(),
        '–ø—Ä–µ–¥–ø—Ä–∏–Ω–∏–º–∞—Ç–µ–ª—å' : df[df['employment_type']=='–ø—Ä–µ–¥–ø—Ä–∏–Ω–∏–º–∞—Ç–µ–ª—å'][column].median(),
        '—Å—Ç—É–¥–µ–Ω—Ç' : df[df['employment_type']=='—Å—Ç—É–¥–µ–Ω—Ç'][column].median(),
        '–≤ –¥–µ–∫—Ä–µ—Ç–µ' : df[df['employment_type']=='–≤ –¥–µ–∫—Ä–µ—Ç–µ'][column].median()
    }
    return dict_fillna
```

```python
# –°—Ç–∞—Å –ú–∏—Ö–∞–π–ª–æ–≤
df.days_employed = df.days_employed.fillna(df.employment_type.map(create_dict_to_fillna('days_employed')))
df.monthly_income = df.monthly_income.fillna(df.employment_type.map(create_dict_to_fillna('monthly_income')))
```

## 3. –û—Ç–≤–µ—Ç—ã –Ω–∞ –≤–æ–ø—Ä–æ—Å—ã –∑–∞–∫–∞–∑—á–∏–∫–∞
* –ï—Å—Ç—å –ª–∏ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å –º–µ–∂–¥—É –Ω–∞–ª–∏—á–∏–µ–º –¥–µ—Ç–µ–π –∏ –≤–æ–∑–≤—Ä–∞—Ç–æ–º –∫—Ä–µ–¥–∏—Ç–∞ –≤ —Å—Ä–æ–∫?
* –ï—Å—Ç—å –ª–∏ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å –º–µ–∂–¥—É —Å–µ–º–µ–π–Ω—ã–º –ø–æ–ª–æ–∂–µ–Ω–∏–µ–º –∏ –≤–æ–∑–≤—Ä–∞—Ç–æ–º –∫—Ä–µ–¥–∏—Ç–∞ –≤ —Å—Ä–æ–∫?
* –ï—Å—Ç—å –ª–∏ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å –º–µ–∂–¥—É —É—Ä–æ–≤–Ω–µ–º –¥–æ—Ö–æ–¥–∞ –∏ –≤–æ–∑–≤—Ä–∞—Ç–æ–º –∫—Ä–µ–¥–∏—Ç–∞ –≤ —Å—Ä–æ–∫?
* –ö–∞–∫ —Ä–∞–∑–Ω—ã–µ —Ü–µ–ª–∏ –∫—Ä–µ–¥–∏—Ç–∞ –≤–ª–∏—è—é—Ç –Ω–∞ –µ–≥–æ –≤–æ–∑–≤—Ä–∞—Ç –≤ —Å—Ä–æ–∫?

–ó–∞–º–µ—Ç–∏–º, —á—Ç–æ –≤—Å–µ –≤–æ–ø—Ä–æ—Å—ã –æ–ø–µ—Ä–∏—Ä—É—é—Ç –∫ —Å—Ç–æ–ª–±—Ü—É `loan_repayment_status`.
–°—Ä–∞–∑—É –Ω–∞–ø–∏—à–µ–º —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è —Å–æ–∑–¥–∞–Ω–∏—è —Å–≤–æ–¥–Ω–æ–π —Ç–∞–±–ª–∏—Ü—ã –ø–æ —ç—Ç–æ–º—É —Å—Ç–æ–ª–±—Ü—É.

```python
def get_repayment_pivot_table(feature_column, aggfunc='count'):
    pivot_table = df.pivot_table(values='loan_purpose',
                                index=[feature_column],
                                columns=['loan_repayment_status'],
                                aggfunc=aggfunc)
    pivot_table['percent'] = pivot_table[1] / (pivot_table[0] + pivot_table[1]) * 100
    pivot_table['percent'] = pivot_table['percent'].round(2)
    return pivot_table
```

```python
def get_plot_by_pivot_table(pivot_table, kind='bar', rot=0):
    bar_figure = pivot_table.plot(y=['percent'], kind=kind, rot=rot)
    return bar_figure
```

```python
quantiles = df['monthly_income'].quantile([0.25, 0.5, 0.75]).tolist()
quantiles
```

```python
def whats_your_income_level(monthly_income, quantiles=quantiles):
    if monthly_income <= quantiles[0]:
            return '–Ω–∏–∑–∫–∏–π'
    elif monthly_income <= quantiles[1]:
            return '–Ω–∏–∑–∫–∏–π —Å—Ä–µ–¥–Ω–∏–π'
    elif monthly_income <= quantiles[2]:
            return '–≤—ã—Å–æ–∫–∏–π —Å—Ä–µ–¥–Ω–∏–π'
    else:
        return '–≤—ã—Å–æ–∫–∏–π'
```

```python
df['income_level'] = df['monthly_income'].apply(whats_your_income_level)

df.sample(10)
```
