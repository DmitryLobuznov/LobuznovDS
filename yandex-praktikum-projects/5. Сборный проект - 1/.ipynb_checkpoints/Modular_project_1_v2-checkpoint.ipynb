## Описание проекта
Вы работаете в интернет-магазине «Стримчик», который продаёт по всему миру компьютерные игры. Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). Вам нужно выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.
Перед вами данные до 2016 года. Представим, что сейчас декабрь 2016 г., и вы планируете кампанию на 2017-й. Нужно отработать принцип работы с данными. Неважно, прогнозируете ли вы продажи на 2017 год по данным 2016-го или же 2027-й — по данным 2026 года.
В наборе данных попадается аббревиатура ESRB (Entertainment Software Rating Board) — это ассоциация, определяющая возрастной рейтинг компьютерных игр. ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».

## 2. Предобработка данных

```python
# Для начала сведём имена столбцов к нижнему регистру.
games.columns = games.columns.str.lower()
games.columns
```

**`year_of_release`, `name`, `genre`**

```python
# Пойдём по порядку проблем, объявленных во время первичного осмотра.
games['year_of_release'] = games['year_of_release'].astype('int32', errors='ignore')
```

```python
# На уровне дальнейней обработки данных отсутствие даты релиза/информации об игре фатально. 
# Доля пропусков мала, не должна сильно повлиять на стат.значения данных.
year_drop_idxes = games[games['year_of_release'].isna()].index
name_drop_idxes = games[games['name'].isna()].index
genre_drop_idxes = games[games['genre'].isna()].index

drop_index = np.unique([*year_drop_idxes, *name_drop_idxes, *genre_drop_idxes])
games = games.drop(drop_index)

# Проверка.
games[['year_of_release', 'name', 'genre']].isna().sum()
```

**`critic_score`**

```python
# Смена типа.
games['critic_score'] = games['critic_score'].astype('int32', errors='ignore')
```

```python
games['critic_score'].isna().sum()
```

**`user_score`**

```python
# Аббревиатура "tbd".
games.query('user_score == "tbd"')
```

```python
# "tbd" = To Be Determined <=> NaN для дальнейшего рассмотрения.
games = games.replace({"user_score" : "tbd"}, np.nan)
```

```python
games['user_score'].isna().sum()
```

```python
games['user_score'] = games['user_score'].astype('float64', errors='ignore')
```

**`rating`**

```python
games['rating'].isna().sum()
```

Пропусков в столбцах `critic_score`, `user_score` и `rating` слишком много, чтобы их было целесообразно удалять => будем работать с пропусками в этих столбцах.  
Отсутствие рейтинга в играх может быть обусловлена либо старостью игры (рейтинг был введён в большинстве стран к 00-ым), либо ошибкой выгрузки. В отношении оценок критиков аналогично. Оценки могли собирать с конкретного ресурса, на котором возможны пробелы в оценках в библиотеке игр.

**Подсчёт суммарных продаж во всех регионах в отдельный столбец**

```python
games['total_sales'] = games.loc[:, 'na_sales': 'other_sales'].values.sum(axis=1)
games.head()
```

*Как менялись продажи по платформам. Выберите платформы с наибольшими суммарными продажами и постройте распределение по годам. За какой характерный срок появляются новые и исчезают старые платформы?*

```python
# Сделаем группировку по платформам.
sales_by_platform = (games
                 .pivot_table(
                     index = 'platform',
                     values = 'total_sales',
                     aggfunc = ['sum']
                 )
                 .sort_values(by=('sum', 'total_sales'), ascending=False)
)
# Возьмём топ 5 платформ для построения распределений по годам.
top_5_sales_by_platform = sales_by_platform.index[:5]
sales_by_platform.head()
```

```python
# Данные по топ 5 платформам.
top_5_platforms = games.query('platform in @top_5_sales_by_platform')
```

```python
# Визуализация распределений.
fig = plt.figure(figsize=(16, 12))

for platform in top_5_sales_by_platform:
    subset = top_5_platforms[top_5_platforms['platform'] == platform]    
    sns.distplot(
        subset['year_of_release'], 
        hist = False,
        kde = True, kde_kws = {'shade' : True, 'linewidth' : 3},
        label = platform
    )
    
plt.legend(prop={'size' : 20}, title = 'Platform')
plt.xlim(left = 1998)
plt.grid()
plt.title('Density Plot with Multiple Platforms', size = 20)
plt.xlabel('Year of release', size = 12)
plt.ylabel('Density', size = 12);
```

**Выводы:**  
Из распределений можем заключить, что характерное время жизни платформы ~`10` лет, а платформы появляются в `5` лет.  
За актуальный период резонно взять срез в 5 лет до прогнозирования.

*Какие платформы лидируют по продажам, растут или падают? Выберите несколько потенциально прибыльных платформ.*
    

```python
# Рейтинг платформ по продажам.
sales_dynamics = (predict_2017
                  .pivot_table(
                      index = ['platform', 'year_of_release'],
                      values = ['total_sales'],
                      aggfunc = ['sum']
                  )
                  .reset_index()
)
sales_dynamics.columns = ['platform', 'year_of_release', 'total_sales']
sales_dynamics.groupby('platform')['total_sales'].sum().sort_values(ascending=False)
```

```python
# Динамика продаж по платформам.
fig = plt.figure(figsize=(16, 12))
palette = sns.color_palette("muted", sales_dynamics['platform'].nunique())
ax = sns.lineplot(
    x = 'year_of_release',
    y = 'total_sales',
    data = sales_dynamics,
    hue = 'platform',
    palette = palette
)
plt.legend(prop={'size' : 15}, title = 'Platform')
ax.set_xlabel('Year of release', size = 12)
ax.set_ylabel('Total sales by year', size = 12)
ax.set_title('Sales Dynamics of platforms by year', size = 20)
ax.grid()
```

**Результат:**  
Динамика на рынке отрицательная по всем платформам.
Однозначным *лидером рынка* является `PS4`.
Второе и третьи места на 2016 год разделили `XOne` и `3DS` соответственно.

*Как влияют на продажи внутри одной популярной платформы отзывы пользователей и критиков. Построение диаграммы рассеяния и посчёт корреляции между отзывами и продажами.*

В качестве популярных платформ возьмём топ 5 (`PS4`, `3DS`, `PS3`, `X360`, `XOne`): 
* Некоторые из платформ видели свой рассвет как раз в период с 2012 по 2016 год, другие только обосновались на рынке;
* На них было выпущено много игр различных жанров;
* Все имеют большой охват пользователей.

```python
platforms_list = ['PS4', '3DS', 'PS3', 'X360', 'XOne']
scores_gravity = games.query('year_of_release >= 2012 & platform in @platforms_list').copy()
scores_gravity.head()
```

```python
fig, axes = plt.subplots(nrows=5, sharex=True, figsize=(20, 30))

for num, platform in enumerate(platforms_list): 
    ax = axes[num]
    sns.scatterplot(
        x = 'critic_score', 
        y = 'total_sales',
        data = scores_gravity[scores_gravity['platform'] == platform],
        size = "total_sales",
        sizes = (25, 150),
        ax = axes[num]
    );
    ax.legend(loc=2)
    ax.grid()
    ax.set_ylabel('Total Sales of %s'%platform, size = 12)
    ax.set_xlabel('Critic scores', size = 12)
    ax.set_title('Influence of Critic scores on Total Sales of %s'%platform, size = 20)
```

```python
fig, axes = plt.subplots(nrows=5, figsize=(20, 30))

for num, platform in enumerate(platforms_list): 
    ax = axes[num]
    sns.scatterplot(
        x = 'user_score', 
        y = 'total_sales',
        data = scores_gravity[scores_gravity['platform'] == platform],
        size = "total_sales",
        sizes = (25, 150),
        ax = ax
    
    );
    ax.legend(loc=2)
    ax.grid()
    ax.set_ylabel('Total Sales of %s'%platform, size = 12)
    ax.set_xlabel('Users scores', size = 12)
    ax.set_title('Influence of Users scores on Total Sales of %s'%platform, size = 20)
```

По диаграммам распределения можем заключить:
* Существует зависимость между оценками критиков и продажами. Для всех популярных платформ есть тенденция к росту после оценки в `70`, как и отсутствие продаж больше 1.5 млн копий при оценке ниже `40`.
* Между оценками пользователей и продажами зависимости замечено не было. Почти для каждой популярной платформы характерно наличие "хита" продаж со средними оценками пользователей.

Проверим наши выводы с помощью коэффициентов корреляции.


```python
corr_slice = scores_gravity[['platform', 'critic_score', 'user_score', 'total_sales']]
fig, axes = plt.subplots(ncols=5, figsize=(16, 6))

for num, platform in enumerate(platforms_list): 
    ax = axes[num]
    sns.heatmap(
        data = corr_slice[corr_slice['platform'] == platform].corr(),
        annot = True,
        cmap = 'coolwarm',
        linewidths=3, linecolor='black', 
        cbar = False,
        ax = axes[num]
    );
    ax.set_title(platform, size = 20)
```

**Результаты:**  
* Коэффициенты корреляции подтвердили прошлый результат касательно зависимости между отзывами и продажами:
    * Отзывы критиков *имеют слабое* влияние на продажи. Сильнее всего это проявляется на платформе `XOne`, слабее - `3DS`;
    * Отзывы игорьков независимы и свободны ото всех оков на всех популярных платформах :D
* Существует весомая зависимость между отзывами критиков и игроков на платформе `3DS`.

## 5. Проверка гипотез

```python
# Гипотезы схожи между собой (равенства ожидаемых средних) => одиннаковый метод проверки => питон же функциональный язык... даже в DataScience :)
def make_ttest_ind(x_1, x_2, alpha = 0.05):
    results = st.ttest_ind(x_1, x_2, equal_var=False)
    print('p-value: ', results.pvalue)
    if (results.pvalue < alpha):
        return "Отвергаем нулевую гипотезу."
    else:
        return "Не отвергаем нулевую гипотезу."
```
