## 2. –ü—Ä–µ–¥–æ–±—Ä–∞–±–æ—Ç–∫–∞ –¥–∞–Ω–Ω—ã—Ö

```python
# –î–ª—è –Ω–∞—á–∞–ª–∞ —Å–≤–µ–¥—ë–º –∏–º–µ–Ω–∞ —Å—Ç–æ–ª–±—Ü–æ–≤ –∫ –Ω–∏–∂–Ω–µ–º—É —Ä–µ–≥–∏—Å—Ç—Ä—É.
games.columns = games.columns.str.lower()
games.columns
```

**`year_of_release`, `name`, `genre`**

* –ú–æ–∂–Ω–æ –∑–∞–º–µ—Ç–∏—Ç—å –∫–∞–∫ —Ä—ã–Ω–æ–∫ –∏–∑–º–µ–Ω–∏–ª—Å—è –ø–æ—Å–ª–µ `2009` –≥–æ–¥–∞. –ú–Ω–æ–≥–æ —Ñ–∞–∫—Ç–æ—Ä–æ–≤ –º–æ–∂–µ—Ç –±—ã—Ç—å: –ø–µ—Ä–µ–Ω–∞—Å—ã—Ç–∏–ª—Å—è, —Å–ª–æ–∂–Ω–æ—Å—Ç—å –ø—Ä–æ–∏–∑–≤–æ–¥—Å—Ç–≤–∞ —É–≤–µ–ª–∏—á–∏–ª–∞—Å—å, —Ç—Ä–µ–±–æ–≤–∞–Ω–∏—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –≤–æ–∑—Ä–∞—Å–ª–∏ –∏ —Ç.–¥. 
____

____
*–ö–∞–∫ –º–µ–Ω—è–ª–∏—Å—å –ø—Ä–æ–¥–∞–∂–∏ –ø–æ –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞–º. –í—ã–±–µ—Ä–∏—Ç–µ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã —Å –Ω–∞–∏–±–æ–ª—å—à–∏–º–∏ —Å—É–º–º–∞—Ä–Ω—ã–º–∏ –ø—Ä–æ–¥–∞–∂–∞–º–∏ –∏ –ø–æ—Å—Ç—Ä–æ–π—Ç–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ –≥–æ–¥–∞–º. –ó–∞ –∫–∞–∫–æ–π —Ö–∞—Ä–∞–∫—Ç–µ—Ä–Ω—ã–π —Å—Ä–æ–∫ –ø–æ—è–≤–ª—è—é—Ç—Å—è –Ω–æ–≤—ã–µ –∏ –∏—Å—á–µ–∑–∞—é—Ç —Å—Ç–∞—Ä—ã–µ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã?*

```python
# –°–¥–µ–ª–∞–µ–º –≥—Ä—É–ø–ø–∏—Ä–æ–≤–∫—É –ø–æ –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞–º.
sales_by_platform = (games
                 .pivot_table(
                     index = 'platform',
                     values = 'total_sales',
                     aggfunc = ['sum']
                 )
                 .sort_values(by=('sum', 'total_sales'), ascending=False)
)
# –í–æ–∑—å–º—ë–º —Ç–æ–ø 5 –ø–ª–∞—Ç—Ñ–æ—Ä–º –¥–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–π –ø–æ –≥–æ–¥–∞–º.
top_5_sales_by_platform = sales_by_platform.index[:5]
sales_by_platform.head()
```

```python
# –î–∞–Ω–Ω—ã–µ –ø–æ —Ç–æ–ø 5 –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞–º.
top_5_platforms = games.query('platform in @top_5_sales_by_platform')
```

```python
# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–π.
fig = plt.figure(figsize=(16, 12))

for platform in top_5_sales_by_platform:
    subset = top_5_platforms[top_5_platforms['platform'] == platform]    
    sns.distplot(
        subset['year_of_release'], 
        hist = False,
        kde = True, kde_kws = {'shade' : True, 'linewidth' : 3},
        label = platform
    )
    
plt.legend(prop={'size' : 20}, title = 'Platform')
plt.xlim(left = 1998)
plt.grid()
plt.title('Density Plot with Multiple Platforms', size = 20)
plt.xlabel('Year of release', size = 12)
plt.ylabel('Density', size = 12);
```

**–í—ã–≤–æ–¥—ã:**  
–ò–∑ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–π –º–æ–∂–µ–º –∑–∞–∫–ª—é—á–∏—Ç—å, —á—Ç–æ —Ö–∞—Ä–∞–∫—Ç–µ—Ä–Ω–æ–µ –≤—Ä–µ–º—è –∂–∏–∑–Ω–∏ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã ~`10` –ª–µ—Ç, –∞ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã –ø–æ—è–≤–ª—è—é—Ç—Å—è –≤ `5` –ª–µ—Ç.

–°–æ–≥–ª–µ—Å–µ–Ω, —Å—É–ø–µ—Ä

> –í–æ–∑—å–º–∏—Ç–µ –¥–∞–Ω–Ω—ã–µ –∑–∞ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏–π –∞–∫—Ç—É–∞–ª—å–Ω—ã–π –ø–µ—Ä–∏–æ–¥. –ê–∫—Ç—É–∞–ª—å–Ω—ã–π –ø–µ—Ä–∏–æ–¥ –æ–ø—Ä–µ–¥–µ–ª–∏—Ç–µ —Å–∞–º–æ—Å—Ç–æ—è—Ç–µ–ª—å–Ω–æ –≤ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–µ –∏—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏—è –ø—Ä–µ–¥—ã–¥—É—â–∏—Ö –≤–æ–ø—Ä–æ—Å–æ–≤. *–û—Å–Ω–æ–≤–Ω–æ–π —Ñ–∞–∫—Ç–æ—Ä ‚Äî —ç—Ç–∏ –¥–∞–Ω–Ω—ã–µ –ø–æ–º–æ–≥—É—Ç –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –ø—Ä–æ–≥–Ω–æ–∑ –Ω–∞ 2017 –≥–æ–¥. –ù–µ —É—á–∏—Ç—ã–≤–∞–π—Ç–µ –≤ —Ä–∞–±–æ—Ç–µ –¥–∞–Ω–Ω—ã–µ –∑–∞ –ø—Ä–µ–¥—ã–¥—É—â–∏–µ –≥–æ–¥—ã.*

```python
predict_2017 = games.query('year_of_release >= 2012')
predict_2017.year_of_release.unique()
```

____
*–ö–∞–∫–∏–µ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã –ª–∏–¥–∏—Ä—É—é—Ç –ø–æ –ø—Ä–æ–¥–∞–∂–∞–º, —Ä–∞—Å—Ç—É—Ç –∏–ª–∏ –ø–∞–¥–∞—é—Ç? –í—ã–±–µ—Ä–∏—Ç–µ –Ω–µ—Å–∫–æ–ª—å–∫–æ –ø–æ—Ç–µ–Ω—Ü–∏–∞–ª—å–Ω–æ –ø—Ä–∏–±—ã–ª—å–Ω—ã—Ö –ø–ª–∞—Ç—Ñ–æ—Ä–º.*
    

```python
# –†–µ–π—Ç–∏–Ω–≥ –ø–ª–∞—Ç—Ñ–æ—Ä–º –ø–æ –ø—Ä–æ–¥–∞–∂–∞–º.
sales_dynamics = (predict_2017
                  .pivot_table(
                      index = ['platform', 'year_of_release'],
                      values = ['total_sales'],
                      aggfunc = ['sum']
                  )
                  .reset_index()
)
sales_dynamics.columns = ['platform', 'year_of_release', 'total_sales']
sales_dynamics.groupby('platform')['total_sales'].sum().sort_values(ascending=False)
```

## 4. –ü–æ—Ä—Ç—Ä–µ—Ç –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è –∫–∞–∂–¥–æ–≥–æ —Ä–µ–≥–∏–æ–Ω–∞

–û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –¥–ª—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è –∫–∞–∂–¥–æ–≥–æ —Ä–µ–≥–∏–æ–Ω–∞ (NA, EU, JP):
____    
*–°–∞–º—ã–µ –ø–æ–ø—É–ª—è—Ä–Ω—ã–µ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã (—Ç–æ–ø-5).*

```python
platforms_pivot = predict_2017.pivot_table(
    index = 'platform',
    values = ['na_sales', 'eu_sales', 'jp_sales', 'total_sales'],
    aggfunc = 'sum'
)
platforms_pivot['na_ratio'] = (platforms_pivot['na_sales'] / platforms_pivot['total_sales']).round(3) 
platforms_pivot['eu_ratio'] = (platforms_pivot['eu_sales'] / platforms_pivot['total_sales']).round(3)
platforms_pivot['jp_ratio'] = (platforms_pivot['jp_sales'] / platforms_pivot['total_sales']).round(3)
platforms_pivot
```

```python
top_5_eu = platforms_pivot[['eu_sales', 'eu_ratio']].sort_values(by='eu_sales', ascending = False).head()
top_5_jp = platforms_pivot[['jp_sales', 'jp_ratio']].sort_values(by='jp_sales', ascending = False).head()
top_5_na = platforms_pivot[['na_sales', 'na_ratio']].sort_values(by='na_sales', ascending = False).head()

print(top_5_eu, top_5_jp, top_5_na)
```
