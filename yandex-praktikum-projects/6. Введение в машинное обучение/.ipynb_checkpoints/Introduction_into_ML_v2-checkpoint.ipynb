## 2. –†–∞–∑–¥–µ–ª–µ–Ω–∏–µ –¥–∞–Ω–Ω—ã—Ö –Ω–∞ –æ–±—É—á–∞—é—â—É—é, –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω—É—é –∏ —Ç–µ—Å—Ç–æ–≤—É—é –≤—ã–±–æ—Ä–∫–∏

–î–∞–Ω–Ω—ã–µ –±—É–¥–µ–º —Ä–∞–∑–¥–µ–ª—è—Ç—å –Ω–∞ 3 —á–∞—Å—Ç–∏: *—Ç—Ä–µ–Ω–∏—Ä–æ–≤–∫–∞*, *–≤–∞–ª–∏–¥–∞—Ü–∏—è* –∏ *—Ç–µ—Å—Ç*, - –ø—Ä–æ–ø–æ—Ä—Ü–∏–æ–Ω–∞–ª—å–Ω–æ $60$% / $20$%  / $20$%.

```python
def train_valid_test_split(df : pd.DataFrame, proportions=[0.6, 0.2, 0.2]):
    """
        Function splits by train, validation, test parts income dataframe.
        In:
            df : pandas.DataFrame() - income dataframe.
            proportions : list() - list of proportions values to split method.
    """
    if sum([*proportions]) != 1:
        raise AttributeError('Wrong "proportions" values. Sum of "proportions" values > 1.0. Must be equal to 1.0.')
    data_list = [train, valid, test] = np.split(
        df.sample(frac=1, random_state=2718281828), 
        [int(proportions[0]*len(df)), int(sum(proportions[:2])*len(df))]
    )
    features, targets = list(), list()
    for data in data_list:
        features.append(data.drop('is_ultra', axis=1))
        targets.append(data['is_ultra'])
    return [*features, *targets]
```

```python
features_train, features_valid, features_test, target_train, target_valid, target_test = \
    train_valid_test_split(df, proportions=[0.6, 0.2, 0.2])
```

```python
table = [
    ['Sample', 'Features_m', 'Features_n', 'Target'],
    ['Train', *features_train.shape, *target_train.shape],
    ['Valid', *features_valid.shape, *target_valid.shape],
    ['Test', *features_test.shape, *target_test.shape]
]
```

```python
def print_table(table):
    longest_cols = [
        (max([len(str(row[i])) for row in table]) + 3)
        for i in range(len(table[0]))
    ]
    row_format = "".join(["{:>" + str(longest_col) + "}" for longest_col in longest_cols])
    for row in table:
        print(row_format.format(*row))
```

```python
print_table(table)
```

## 3. –ò—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏–µ –∫–∞—á–µ—Å—Ç–≤–∞ –º–æ–¥–µ–ª–µ–π –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤

**–í—ã–±—Ä–∞–Ω–Ω—ã–µ –º–æ–¥–µ–ª–∏:**
* *–†–µ—à–∞—é—â–µ–µ –¥–µ—Ä–µ–≤–æ* (**DecisionTreeClassifier**);
* *–°–ª—É—á–∞–π–Ω—ã–π –ª–µ—Å* (**RandomForestClassifier**);
* *–õ–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–∞—è —Ä–µ–≥—Ä–µ—Å—Å–∏—è* (**LogisticRegression**);
* *–ú–µ—Ç–æ–¥ $k$ –ë–ª–∏–∂–∞–π—à–∏—Ö –°–æ—Å–µ–¥–µ–π* (**KNeighborsClassifier**);
* *–ù–∞–∏–≤–Ω—ã–π –ë–∞–π–µ—Å–æ–≤—Å–∫–∏–π –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ç–æ—Ä* (**GaussianNB**);
* *–ú–µ—Ç–æ–¥ –æ–ø–æ—Ä–Ω—ã—Ö –≤–µ–∫—Ç–æ—Ä–æ–≤* (**SVC**).
____

```python
# –†–∞–∑–º–µ—Ä –≤—ã–±–æ—Ä–æ–∫.
features_train.shape, target_train.shape
```

**`–†–µ—à–∞—é—â–µ–µ –¥–µ—Ä–µ–≤–æ (DecisionTreeClassifier)`** 

```python
best_accuracy = 0
for depth in range(1, 21):
    model = DecisionTreeClassifier(max_depth=depth, random_state=2718281828)
    model.fit(features_train, target_train)
    predictions_valid = model.predict(features_valid)
    accuracy = accuracy_score(target_valid, predictions_valid)
    if accuracy > best_accuracy:
        best_accuracy = accuracy
        print('best_depth =', depth, ':', best_accuracy)
        decision_tree_model = model
decision_tree_model
```

**`–°–ª—É—á–∞–π–Ω—ã–π –ª–µ—Å (RandomForestClassifier)`**

```python
best_accuracy = 0
for estimator in range(1, 21):
    for depth in range(1, 21):
        model = RandomForestClassifier(n_estimators=estimator, max_depth=depth, random_state=2718281828)
        model.fit(features_train, target_train)
        predictions_valid = model.predict(features_valid)
        accuracy = accuracy_score(target_valid, predictions_valid)
        if accuracy > best_accuracy:
            best_accuracy = accuracy
            print('n_estimators =', estimator, '& best_depth =', depth, ':', best_accuracy)
            random_forest_model = model
random_forest_model
```

<div class="alert alert-block alert-success">

**`–õ–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–∞—è —Ä–µ–≥—Ä–µ—Å—Å–∏—è (LogisticRegression)`**

```python
best_accuracy, best_max_iter = 0, 0
for iter in range(100, 1100, 100):
    model = LogisticRegression(max_iter=iter, random_state=2718281828, solver='lbfgs')
    model.fit(features_train, target_train)
    predictions_valid = model.predict(features_valid)
    accuracy = accuracy_score(target_valid, predictions_valid)
    if accuracy > best_accuracy:
        best_accuracy = accuracy
        print('best_max_iter =', iter, ':', best_accuracy)
        log_regr_model = model
log_regr_model
```

```python
best_accuracy = 0
for num in range(5, 21):
    for p in range(2, 5):
        model = KNeighborsClassifier(n_neighbors=num, p=p)
        model.fit(features_train, target_train)
        predictions_valid = model.predict(features_valid)
        accuracy = accuracy_score(target_valid, predictions_valid)
        if accuracy > best_accuracy:
            best_accuracy = accuracy
            print('best_n_neighbors =', num, '& dim Minkowski space =', p, ':', best_accuracy)
            knn_model = model
knn_model
```
