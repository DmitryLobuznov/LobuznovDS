```python
import pandas as pd
import numpy as np
from dateutil.parser import parse
```

## Произведём закачку найденных данных с `Investing.com`

```python
tickers = pd.read_csv(r'C:\Users\Дмитрий\Desktop\Watchlist_18022021.csv')
tickers.head()
```

```python
tickers.columns = ['name', 'ticker', 'price', 'open', 'max', 'min', 'change', 'change_percent', 'volume', 'date_time']
tickers.head()
```

```python
tickers = tickers.drop(['change_percent'], axis=1)
tickers.head()
```

```python
tickers.info(), tickers.isna().sum()
```

```python
# Float columns
float_cols = ['price', 'open', 'max', 'min', 'change']
for col in float_cols:
    tickers[col] = tickers[col].apply(lambda x: float(x.split()[0].replace(',', '').replace('.', '')))
tickers[float_cols] = tickers[float_cols].astype(float)
tickers.head()
```

```python
tickers.date_time.apply(lambda x: x.astype('datetime'))
```
